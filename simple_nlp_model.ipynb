```python
from nltk.tokenize import word_tokenize ,sent_tokenize

paragraph = "we love python very much. I am a student of computer science and engineering and i also know about and love javascript"

tokenized_word = word_tokenize(paragraph)
tokenized_sentence = sent_tokenize(paragraph)

print("tokenized_word :", tokenized_word)
print("tokenized_sentence : ", tokenized_sentence)
```

```python
# Frequency Distribution 
from nltk.probability import FreqDist

freq_dist = FreqDist(tokenized_word)

print("freq_dist :", freq_dist)

freq_dist.most_common(2)
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt

freq_dist.plot(30,cumulative=False)

plt.show()
```

```python
# stopwords
# import nltk
# nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

print(stop_words)
```

```python
# Removing Stopwords 
filtered_sentence=[]

for w in tokenized_word:
    
    if w not in stop_words:
        
        filtered_sentence.append(w)
        
print("Tokenized Sentence:",tokenized_word)
print("Filterd Sentence:",filtered_sentence)
```
