```python
from nltk.tokenize import word_tokenize ,sent_tokenize

paragraph = """we love python very much. I am a student of computer science and engineering and i also know about and love javascript. connecting to the internet. 
good at programming.he is better than you."""

tokenized_word = word_tokenize(paragraph)
tokenized_sentence = sent_tokenize(paragraph)

print("tokenized_word :", tokenized_word)
print("tokenized_sentence : ", tokenized_sentence)
```

```python
# Frequency Distribution 
from nltk.probability import FreqDist

freq_dist = FreqDist(tokenized_word)

print("freq_dist :", freq_dist)

freq_dist.most_common(2)
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt

freq_dist.plot(30,cumulative=False)

plt.show()
```

```python
# stopwords
# import nltk
# nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

print(stop_words)
```

```python
# Removing Stopwords 
filtered_sentence=[]

for w in tokenized_word:
    
    if w not in stop_words:
        
        filtered_sentence.append(w)
        
print("Tokenized Sentence:",tokenized_word)
print("Filterd Sentence:",filtered_sentence)
```

```python
# Stemming
from nltk.stem import PorterStemmer

ps = PorterStemmer()

stemmed_words=[]

for w in filtered_sentence:
    stemmed_words.append(ps.stem(w))

print("Filtered Sentence:",filtered_sentence)
print("Stemmed Sentence:",stemmed_words)
```

```python
#performing stemming and Lemmatization
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer

lem = WordNetLemmatizer()
stem = PorterStemmer()

word = "flying"
print("Lemmatized Word:",lem.lemmatize(word,"v"))
print("Stemmed Word:",stem.stem(word))
```

```python
# POS tag 

import nltk
nltk.pos_tag(filtered_sentence)
```
